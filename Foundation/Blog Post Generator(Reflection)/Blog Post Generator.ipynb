```python
import os
import anthropic
```

```python
# Set your API key: 
os.environ["ANTHROPIC_API_KEY"] = "your api key here"
client = anthropic.Anthropic()
```

```python
def generate_post(topic: str, audience: str) -> str:
    """Producer: write a first draft"""
    r = client.messages.create(
        model="claude-sonnet-4-5", max_tokens=600,
        system=f"You are a content writer for {audience}. Write clearly and concisely.",
        messages=[{"role": "user", "content":
            f"Write a 300-word blog post intro about: {topic}"}])
    return r.content[0].text
```

```python
def critique_post(draft: str, audience: str) -> str:
    """Critic: evaluate against specific criteria(SEPARATE call — key architectural point)"""
    r = client.messages.create(
        model="claude-sonnet-4-5", max_tokens=400,
        system="You are a harsh but fair content editor. Be specific about what is weak.",
        messages=[{"role": "user", "content":
            f"""Review this blog post for {audience}. Evaluate:
1. Opening hook strength(is it compelling?)
2. Clarity of the core message
3. Specific examples vs. vague claims
4. Actionability — does the reader know what to do?
5. Tone match for {audience}

Draft:
{draft}

List concrete improvements. Be specific."""}])
    return r.content[0].text
```

```python
def revise_post(draft: str, critique: str) -> str:
    """Producer: revise using critic's feedback"""
    r = client.messages.create(
        model="claude-sonnet-4-5", max_tokens=600,
        messages=[{"role": "user", "content":
            f"Rewrite this post incorporating all the feedback:\n\nOriginal:\n{draft}\n\nFeedback:\n{critique}\n\nRevised version:"}])
    return r.content[0].text
```

```python
topic    = "why product managers need to understand AI agents"
audience = "senior product managers at Indian tech companies"
```

```python
draft    = generate_post(topic, audience)
critique = critique_post(draft, audience)
revised  = revise_post(draft, critique)

print("=== DRAFT ===");   print(draft)
print("\n=== CRITIQUE ==="); print(critique)
print("\n=== REVISED ===");  print(revised)
```
